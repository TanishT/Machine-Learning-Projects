```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import misc
i = misc.ascent() #getting image from scipy library
```

```python
i_transformed = np.copy(i) #cloning array i
size_x = i_transformed.shape[0]
size_y = i_transformed.shape[1]
```

```python
#pooling algorithm - 2 by 2 pooling filter
new_x = int(size_x/2)
new_y = int(size_y/2)
newImage = np.zeros((new_x, new_y))

#third variable is 2 because the stride for the pooling algorithm is 2
for x in range(0, size_x, 2):
    for y in range(0, size_y, 2):
        pixels = []
        pixels.append(i_transformed[x, y])
        pixels.append(i_transformed[x+1, y])
        pixels.append(i_transformed[x, y+1])
        pixels.append(i_transformed[x+1, y+1])
        newImage[int(x/2),int(y/2)] = max(pixels)

# Plot the image. Note the size of the axes -- now 256 pixels instead of 512
plt.gray()
plt.grid(False)
plt.imshow(newImage) #original image was 512 by 512, now image is 256 by 256. 1/2 of the size but same quality
plt.show()  
```
